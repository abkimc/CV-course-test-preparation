```python
#1 When is a classification approach preferable, and when should regression be chosen?

"""
Classification tasks aim to assign a label or category to an entire image or specific
objects within an image. The goal is to categorize data into discrete classes.
In image processing, regression tasks might involve predicting the
coordinates of objects within an image, estimating the age of individuals from their photos, or
any other task where the output is a continuous value.
"""

#2 How are supervised and unsupervised learning distinguished from each other?

"""
Supervised learning: In this type of learning, we are given
an input dataset along with the correct labels, and we
need to learn the relationship (as a function) between the
input and the output. The handwritten-digit classification
problem is an example of a supervised (classification)
problem.

Unsupervised learning: In this type of learning, we have little
or no idea what our output should look like. We can derive
structure from data where we dont necessarily know the
effect of the variables. An example is clustering, which can
also be considered segmentation, in image processing
techniques where we do not know which pixel belongs to
which segment.
"""

#3 What models are included under unsupervised machine learning, and which are considered supervised machine learning models?

"""
Supervised Models
 Linear Regression
 Lasso Regression
 Ridge Regression
 All tree models
 SVM
 KNN

"""

#4 In what ways do object recognition and image classification differ?

"""
Image Classification: This task identifies the images dominant object or
main theme. For instance, given an image containing a dog, the goal of
image classification would be to label the image as "dog".

Object recognition, sometimes called object detection, involves identifying
all objects in an image and classifying them into known categories. It further
determines the objects location within the image, often represented by
bounding boxes or more precise segmentation masks.

"""

#5 What are the pros and cons of Nearest Mean Classification?

"""

Pros
   Simplicity: The Nearest Mean Classifier is straightforward to understand and implement. It doesn't
  require complex iterative training like other algorithms like neural networks or support vector
  machines.
   Interpretability: The classification decision is based on clear and interpretable criteriathe
  distance to class centroids.
   Scalability: Since the model complexity does not grow with the training data size (only one
  centroid per class is stored), the classifier is relatively scalable and efficient regarding memory
  usage.

Cons
   Sensitivity to Outliers: The mean is sensitive to outliers, which can skew the centroid and potentially
  degrade classification performance.
   Assumption of Class Similarity: The classifier implicitly assumes all classes have similar variance and
  shape in the feature space. If one class is more spread out than another, the classifier may be
  biased towards the more compact classes.
   Linear Decision Boundaries: The decision boundaries between classes are linear, which may not
  be adequate for datasets where classes are not linearly separable.
"""

#6 What varieties of linear classification methods exist?

"""
 SVM
 Logistic Regression
 Linear Discriminant Analysis (LDA)

"""

#7 In the context of regression, what is meant by One-Hot Encoding?

"""
One-hot encoding is a process used to convert categorical variables into a form
that ML algorithms can use to improve prediction. It creates a binary column for
each variable category and is particularly useful in regression analysis when
dealing with categorical predictors.
"""

#8 Under what circumstances is Polynomial Fitting applied to linear problems?

"""
using polynomial fitting within a linear classification
framework lies in the ability to model complex, non-linear relationships between
features and the target variable while still employing linear models. This approach
leverages the power of linear models to handle non-linear data by transforming the
feature space.
"""

#9 What is considered overfitting in classification problems?

"""
Overfitting in classification problems is a common issue where a model learns the
detail and noise in the training data to the extent that it negatively impacts the
model's performance on new, unseen data. An overfitted model needs to be
simplified, capturing spurious patterns in the training data that do not generalize to
other data sets. This results in high accuracy on the training data but needs to
improve on the test data or any new data it encounters.
"""

#10 Which strategies can be implemented to prevent overfitting?

"""
Cross-Validation: Use cross-validation techniques to ensure the model's performance is consistent across different
data subsets, helping ensure the model generalizes well.

Simplify the Model: Reduce the complexity of the model by selecting fewer parameters or features, using simpler
model architectures, or pruning decision trees.

Regularization: Apply regularization techniques, which add a penalty on the size of coefficients for the loss
function to discourage learning a more complex or flexible model.

Increase Training Data: More data can help the model learn the signal better and reduce the noise's impact,
though this might only sometimes be feasible.

Feature Selection: Carefully select the most relevant features for the model, reducing the chance of learning from
noise in the data.

Early Stopping: In iterative models like neural networks, stop training before the model overfits by monitoring the
performance on a validation set and stopping when performance degrades.

Ensemble Methods: Techniques like bagging and boosting can help reduce overfitting by averaging biases and
reducing variance.
"""

#11 What are the differences between PCA (Principal Component Analysis) and LDA (Linear Discriminant Analysis)?

"""
PCA: Maximize the projected variance
LDA: Maximize between classes, minimize within-class variance

 Supervision: PCA is unsupervised, not considering class labels, whereas
LDA is supervised, using class labels to guide the dimensionality reduction
process.
 Objective: PCA focuses on capturing the directions of maximum variance,
regardless of class labels. LDA aims to maximize class separability.
 Assumptions: LDA makes stronger assumptions about the data distribution,
including normalization and equal class covariances, which do not apply
to PCA.
"""

#12 How does a Random Forest algorithm function?

"""
1. Bootstrap Aggregating (Bagging):
Random Forests use the bagging technique to create multiple subsets of the original
training data. For each subset, a decision tree is trained

2. Feature Randomness:
When building each tree, instead of searching for the best feature when splitting a node, the
algorithm searches for the best feature among a random
subset of features. This adds an extra layer of randomness to
the model, which helps increase diversity among the trees
and results in a more robust overall model.

3. Building Multiple Decision Trees:
Many decision trees are constructed. Each tree is built to the maximum depth
possible, and no pruning is done. The idea is that individual
trees might overfit different parts of the training data, but
their errors will cancel out when averaged.

4. Majority Voting for Classification:
Each decision tree in the forest outputs a class prediction for a classification task, and
the final output prediction is based on a majority vote. That is,
the predicted class is the one that gets the most votes from
the individual trees.
"""

#13 What is the operation mechanism for SVM (Support Vector Machine)?

"""
1. Maximizing the Margin: The algorithm looks for the hyperplane with the maximum margin,
i.e., the maximum distance between data points of both classes. Maximizing the margin
increases the model's robustness and ability to generalize well to unseen data.

2.Handling Non-Linearly Separable Data: For datasets that are not linearly separable, SVM
uses a technique called the kernel trick. The kernel trick involves mapping the input
features into a higher-dimensional space where a linear separation is possible. Common
kernels include the linear, polynomial, Radial Basis Function (RBF), and sigmoid.

3. Soft Margin Classification: To handle the possibility of data overlap and to improve the
model's generalization capabilities, SVM introduces the concept of the soft margin. This
approach allows some points to be on the wrong side of the margin or the hyperplane,
controlled by a regularization parameter (often denoted as C). A smaller value of C leads
to a wider margin and more margin violations but might improve the models
generalization ability.

4. Optimization Problem: Finding the optimal hyperplane is a convex optimization problem
that can be efficiently solved using sequential minimal optimization (SMO) techniques.
"""

#14 Why is the confusion matrix important, and what does it represent?

"""
It is a performance
measurement for a machine
learning classification problem
where the output can be two or
more classes.
It is a table with 4 different
predicted and actual value
combinations.

 True Positive (TP): The total counts
having predicted, and actual values
are Dog.
 True Negative (TN): It is the total
counts having both predicted and
actual values are Not Dog.
 False Positive (FP): It is the total counts
having prediction is Dog while actually
Not Dog.
 False Negative (FN): It is the total
counts having prediction is Not Dog
while actually, it is Dog.
"""

#15 How is a Confusion Matrix calculated for a binary classification problem?

"""
regualerly
"""

#16 What distinguishes machine learning classification from deep learning classification?

"""
Model Complexity:

Machine Learning: Typically involves simpler models like Support Vector Machines (SVM), Decision
Trees, or shallow neural networks. These models can perform well on image classification tasks but
might struggle with complex patterns and high-dimensional data without carefully crafted
features.

Deep Learning: Utilizes more complex models, such as Convolutional Neural Networks (CNNs),
specifically designed to process grid-like data like images. CNN, through its deep architecture,
can capture hierarchical patterns for more accurate classification.

Data Requirements:

Machine Learning: Can often work with smaller datasets due to the reliance on manually
engineered features designed to capture the essence of the data.

Deep Learning: Generally requires large amounts of labeled data to learn features and minimize
overfitting effectively. The performance of deep learning models typically improves with the
amount of data available.
ML and DL classification

Computational Resources:

 Machine Learning: Less computationally
intensive than deep learning, making it
possible to train and deploy models on less
powerful hardware.

 Deep Learning requires significant
computational resources, especially GPUs, for
training because of its complexity and depth.
Training deep learning models can take
considerable time and computational power.

"""
```
